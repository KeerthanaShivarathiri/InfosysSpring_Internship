```python
#  Analyse priority and queue

import matplotlib.pyplot as plt
import seaborn as sns

# Assuming 'priority' and 'queue' are categorical columns
# Create a cross-tabulation of priority and queue counts
priority_queue_counts = pd.crosstab(df['priority'], df['queue'])

# Create the heatmap with red color shades
plt.figure(figsize=(10, 6))
sns.heatmap(priority_queue_counts, annot=True, fmt="d", cmap="Reds")
plt.title("Heatmap of Priority vs. Queue")
plt.xlabel("Queue")
plt.ylabel("Priority")
plt.show()
```

```python
# prompt:  Analyse priority and business type

# Analyze priority and business_type

# Assuming 'priority' and 'business_type' are categorical columns
# Create a cross-tabulation of priority and business_type counts
priority_business_counts = pd.crosstab(df['priority'], df['business_type'])

# Create the heatmap with different color shades (e.g., Blues)
plt.figure(figsize=(12, 6))  # Adjust figure size if needed
sns.heatmap(priority_business_counts, annot=True, fmt="d", cmap="Blues")
plt.title("Heatmap of Priority vs. Business Type")
plt.xlabel("Business Type")
plt.ylabel("Priority")
plt.show()
```

```python
# prompt:  Relation between type and queue

# Analyze the relationship between 'type' and 'queue'

# Create a cross-tabulation of type and queue counts
type_queue_counts = pd.crosstab(df['type'], df['queue'])

# Create the heatmap
plt.figure(figsize=(12, 6))  # Adjust figure size as needed
sns.heatmap(type_queue_counts, annot=True, fmt="d", cmap="YlGnBu") # Use a different colormap
plt.title("Heatmap of Type vs. Queue")
plt.xlabel("Queue")
plt.ylabel("Type")
plt.show()

```
